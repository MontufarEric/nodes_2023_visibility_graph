```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

# Summarize metrics
def summarize_metrics(data_list):
    metrics = {}
    for metric in data_list[0].keys():
    # if isinstance(metrics[metric],int):
        metrics[metric] = [item[metric] for item in data_list  if isinstance(item[metric],(int,float))]
    return metrics
```

```python
# Load JSON files
with open('D:/Documents/Projects/Nodes/SanosDormidos_diclist_test.json', 'r') as f:
    sanos = json.load(f)

with open('D:/Documents/Projects/Nodes/CongestivossDormidos_dic.json', 'r') as f:
    cong = json.load(f)
```

```python
sanos_sum = summarize_metrics(sanos)
cong_sum = summarize_metrics(cong)
```

```python
sanos_sum = summarize_metrics(sanos)
cong_sum = summarize_metrics(cong)

# Compare and plot
metrics_names = list(sanos_sum.keys())
avg_values1 = [sanos_sum[metric] for metric in metrics_names]
avg_values2 = [cong_sum[metric] for metric in metrics_names]

x = range(len(metrics_names))

plt.figure(figsize=(15, 7))
plt.bar(x, avg_values1, width=0.4, label='sanos', color='b', align='center')
plt.bar(x, avg_values2, width=0.4, label='cong', color='r', align='edge')
plt.xlabel('Metrics')
plt.ylabel('Average Value')
plt.title('Comparison of Metrics between List 1 and List 2')
plt.xticks(x, metrics_names, rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
summary1 = summarize_metrics(sanos)
summary2 = summarize_metrics(cong)

# Number of metrics
num_metrics = len(summary1)

# Create subplots for each metric
fig, axs = plt.subplots(num_metrics, 1, figsize=(10, 5*num_metrics))

for i, metric in enumerate(summary1.keys()):
    axs[i].bar(['List 1', 'List 2'], [summary1[metric], summary2[metric]], color=['b', 'r'])
    axs[i].set_ylabel('Average Value')
    axs[i].set_title(metric)

plt.tight_layout()
plt.show()
```

```python
def extract_degrees(data_list):
    degrees = []
    for item in data_list:
        degrees.extend([entry["degree"] for entry in item["node_degrees"]])
    return degrees
```

```python
# Extracting degrees from both lists
degrees_list1 = extract_degrees(sanos)
degrees_list2 = extract_degrees(cong)
# Calculating the degree distributions
degree_values1, degree_counts1 = np.unique(degrees_list1, return_counts=True)
degree_values2, degree_counts2 = np.unique(degrees_list2, return_counts=True)

degree_counts1 = degree_counts1 / len(degrees_list1)
degree_counts2 = degree_counts2 / len(degrees_list2)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(degree_values1, degree_counts1, width=0.4, label='Sanos', align='center')
plt.bar(degree_values2, degree_counts2, width=0.4, label='cong', align='edge')
plt.xlabel('Degree')
plt.ylabel('Probability')
plt.title('Comparison of Degree Distributions')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 20 random data points
data = np.random.randn(20)

# Calculate mean and standard deviation
mean_val = np.mean(data)
std_val = np.std(data)

# Plotting the time series data
plt.figure(figsize=(10,6))
plt.plot(data, '-o', label="Data Points")
plt.axhline(y=mean_val, color='blue', linestyle='-', label="Mean")
plt.axhline(y=mean_val + std_val, color='red', linestyle='--', label="Mean + 1 Std")
plt.axhline(y=mean_val - std_val, color='red', linestyle='--', label="Mean - 1 Std")

# Adding legends, title, axis labels and adjusting font size for clarity
plt.title("Time Series with Mean and Standard Deviation", fontsize=16)
plt.xlabel("Time", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.xticks(np.arange(0, 21),fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_bars_with_lines(connection_dict):
    # Number of bars
    n_bars = len(connection_dict.keys())
    
    # Random data and colors for each bar
    
    colors = plt.cm.viridis(np.linspace(0, 1, n_bars))

    # Plotting the data as bars with thinner bars and different colors
    plt.figure(figsize=(10,6))
    bars = plt.bar(connection_dict.keys(), data, color=colors, width=0.2)

    # Creating lines between bars
    for bar, targets in connection_dict.items():
        x1 = bar
        y1 = data[bar-1]  # Minus 1 since array is 0-indexed
        color = colors[bar-1]  # Use the color of the current bar
        
        for target in targets:
            x2 = target
            y2 = data[target-1]
            
            x_vals = [bars[x1-1].get_x() + bars[x1-1].get_width() / 2, 
                      bars[x2-1].get_x() + bars[x2-1].get_width() / 2]
            y_vals = [y1, y2]
            
            plt.plot(x_vals, y_vals, color=color, linewidth=2)

    # Adding title, axis labels and adjusting font size for clarity
    # plt.title("Bar Chart with Lines", fontsize=16)
    plt.xlabel("Time", fontsize=14)
    plt.ylabel("Value", fontsize=14)
    plt.xticks(list(connection_dict.keys()), fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, axis='y')
    plt.tight_layout()

    # Show the plot
    plt.show()

# Example usage
connections = {
    1: [2,3],
    2: [3],
    3: [4,5,6],
    4: [5,6],
    5: [6],
    6: [7],
    7: []
}
plot_bars_with_lines(connections)
```
